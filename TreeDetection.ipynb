```python
import torch
import cv2
from models.experimental import attempt_load
from utils.general import non_max_suppression, scale_coords
from utils.plots import plot_one_box

# Load the YOLOv8 model
weights_path = r"C:\Users\BHEEMA RAJU\runs\detect\yolov8n_custom12\weights\best.pt"
model = attempt_load(weights_path, map_location=torch.device('cuda' if torch.cuda.is_available() else 'cpu'))
stride = int(model.stride.max())  # Model stride

# Set the device to GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
if device == 'cuda':
    model.half()

# Load the video
video_path = r"C:\Users\BHEEMA RAJU\Downloads\190416_10_Drone1_08_preview.mp4"
cap = cv2.VideoCapture(video_path)

# Loop over the frames
while cap.isOpened():
    # Read a frame
    ret, frame = cap.read()
    if not ret:
        break

    # Resize the frame to match the YOLO input size
    img = cv2.resize(frame, (640, 640))

    # Convert BGR to RGB
    img = img[:, :, ::-1].transpose(2, 0, 1)

    # Normalize and convert to torch tensor
    img = torch.from_numpy(img).to(device)
    img = img.half() if device == 'cuda' else img.float()
    img /= 255.0
    if img.ndimension() == 3:
        img = img.unsqueeze(0)

    # Perform inference
    pred = model(img)[0]

    # Apply non-maximum suppression
    pred = non_max_suppression(pred, 0.4, 0.5)

    # Process detections and draw bounding boxes
    for det in pred[0]:
        if det is not None and len(det):
            det[:, :4] = scale_coords(img.shape[2:], det[:, :4], frame.shape).round()
            for *xyxy, conf, cls in reversed(det):
                label = f'{int(cls)}: {conf:.2f}'
                plot_one_box(xyxy, frame, label=label, color=(0, 255, 0), line_thickness=2)

    # Show the frame
    cv2.imshow('YOLOv8', frame)

    # Press Q to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and destroy all windows
cap.release()
cv2.destroyAllWindows()
```
